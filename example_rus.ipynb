# Работа с формами отчётности

Далее выгрузим все имеющиеся в формы отчётности в pandas формате. 

Импортируем соответствующие классы:

```python
from banks.forms import Form101, Form102, Form123, Form135
```

Каждый из импортируемых классов представляет структуру формы и позволяет получать информацию о её основных разделах (аттрибутах класса). На примере формы 101:

```python
bank = Bank(350000004,1481,'Сбербанк России')
form101 = Form101(bank)
form101.struct.head()
```

Строки обозначают отдельные счета и информацию об их классификации - балансовый/внебалансовый, актив/пассив и т.д.. Например, получить только вклады физ. лиц можно следующим образом:

```python
form101.balance_accounts.\
Депозиты_и_прочие_привлеченные_средства_физических_лиц.accounts
```

При инициализации форма не содержит в себе каких-либо реальных данных, кроме структуры. Чтобы скачать данные воспользуемся методом fill(), которые в качестве аргумента принимает количество форм, которые необходимо загрузить, начиная с последней отчётной даты:

```python
form101.fill(4)
```

После заполнения, форма будет содержать в себе даты и остатки на счетах. Для иллюстрации вернёмся к предыдущему примеру:

```python
print(form101.balance_accounts.\
      Депозиты_и_прочие_привлеченные_средства_физических_лиц.\
      liabilities_sum())
# Аналогично кредитный портфель
print(form101.balance_accounts.\
      Кредиты_и_прочие_средства__предоставленные_физическим_лицам.\
      assets_sum())
```

У формы так же есть метод to_dataframe(), преобразующий её (или её часть) в pandas формат:

```python
form101.to_dataframe().head()
```

Работа с другими формами осуществляется аналогичным образом. Теперь можно выгрузить ВСЕ имеющиеся 101-е формы :) У меня это заняло довольно много времени.

```python
from multiprocessing import Pool, cpu_count
import functools
import pandas as pd

# Объединим вместе открытые и закрытые
all_banks = bs.closed_banks_list + bs.active_banks_list

def load_form(bank, form = Form101, first_n = None):
    print("{0} - {1}".format(bank.bank_id, bank.name))
    try:
        return form(bank).fill(first_n).to_dataframe()
    except:
        return pd.DataFrame()
```

```python
with Pool(processes=cpu_count()) as pool:
    results = pool.map(
        load_form,
        all_banks,
    )
f_101 = pd.concat(results)
f_101 = pd.concat(results)
f_101.to_csv('101f.csv', index=False)
```

```python
with Pool(processes=8) as pool:
    results = pool.map(
        functools.partial(load_form, form = Form102),
        all_banks,
    )
f_102 = pd.concat(results)
f_102.to_csv('102f.csv', index=False)
```

```python
with Pool(processes=8) as pool:
    results = pool.map(
        functools.partial(load_form, form = Form123),
        all_banks,
    )
f_123 = pd.concat(results)
f_123.to_csv('123f.csv', index=False)
```

Для 134 формы используется тот же класс, что и для 123, с специальным параметром типа формы - 'f_134'

```python
with Pool(processes=8) as pool:
    results = pool.map(
        functools.partial(load_form, form = functools.partial(Form123,form_type='f_134')),
        all_banks,
    )
f_134 = pd.concat(results)
f_134.to_csv('134f.csv', index=False)
```

```python
with Pool(processes=8) as pool:
    results = pool.map(
        functools.partial(load_form, form = Form135),
        all_banks,
    )
f_135 = pd.concat(results)
f_135.to_csv('135f.csv', index=False)
```

```python
from multiprocessing import Pool, cpu_count
import functools
import pandas as pd

all_banks = bs.closed_banks_list + bs.active_banks_list

def get_info(bank):
    try:
        print("{0} - {1}".format(bank.bank_id, bank.name))
        return bank.get_main_info()
    except ValueError:
        return pd.DataFrame()

with Pool(processes=8) as pool:
    results = pool.map(
        get_info,
        all_banks,
    )
main_info = pd.concat(results)
main_info.to_csv('main_info.csv', index=False)
```
