```python
def get_closed_banks(url):
    page = urlopen(url).read()
    closed_banks = pd.read_html(page)[2]
    closed_banks['link'] = pd.Series([a['href'] for a in BeautifulSoup(page, 'html.parser').\
                                   findAll('table')[2].find('tbody').findAll('a')])
    return closed_banks

def get_definition_list(url):
    page = urlopen(url).read()
    bank_definition = BeautifulSoup(page,'html.parser').find('dl')
    definition = pd.DataFrame(data = [' '.join(dt.text.split()) for dt in bank_definition.findAll('dd')],
                              index=[' '.join(dt.text.split()) for dt in bank_definition.findAll('dt')]).T
    definition['cbr_link'] = bank_definition.find('a')['href']
    
    definition['Номер лицензии'] = definition['Номер лицензии'].map(lambda x: x.split()[0])
    
    definition.columns = ['full_name', 'city', 
                          'license_number', 'reason_of_closing', 'date_of_closing',
                          'reason_extended', 'cbr_link']
    return definition
```

```python
first_page = 'http://www.banki.ru/banks/memory/'

all_pages = [(first_page + '?PAGEN_1=%d') % x for x in range(2,51)]

closed_banks = get_closed_banks(first_page)

counter = 1

for page_url in all_pages:
    print('\rProcessed %d of %d...' % (counter, len(all_pages)+1), end='')
    closed_banks = pd.concat([closed_banks, get_closed_banks(page_url)])
    
    counter+=1
    
closed_banks.columns = ['index', 'bank', 'license_number', 'reason_of_closing', 'date_of_closing', 'city','link']
```

```python
bank_definitions = pd.DataFrame()

for bank in closed_banks.head():
    url = 'http://www.banki.ru/'+ bank[1]['link']
    bank_definitions = pd.concat([bank_definitions,get_definition_list(url)])
```

```python


def get_f_135(f_135):
    start_date = "2012-01-01"

    urls = [('http://www.cbr.ru/credit/'+a['href']).replace('®','&reg') for a in f_135.findAll('a')]
     #       if len('http://www.cbr.ru/credit/'+a['href']) == 63 ]
    forms_135 = pd.DataFrame(dict(url=urls,
                              date=[dtparser.parse(url[url.find('when=')+5:url.find('when=')+13]) 
                                    for url in urls])).query('date > @start_date')
    f_135 = pd.DataFrame()

    for form in forms_135.iterrows():

        page = urlopen(form[1]['url']).read()
        tables = pd.read_html(page)

        transcripts = tables[1].drop(0)
        norms = tables[2][[0,1]].drop(0)

        ul = BeautifulSoup(page, 'html.parser').find('ul', {'class':'without_dash without_indent'})
        indicators = []
        for li in ul.findAll('li'):
            li = ''.join(str(li.text).split())
            indicators.append([li[:li.find('=')],li[li.find('=')+1:]])

        indicators = pd.DataFrame(indicators)

        table = pd.concat([transcripts, norms, indicators]).set_index(0).fillna(0)
        table[1] = table[1].astype(np.str).map(to_number)
        table = table.T
        table['date'] = form[1]['date']

        f_135 = pd.concat([f_135, table])
        
    return f_135
```
